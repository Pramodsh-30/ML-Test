```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, accuracy_score

# Drop rows with missing values
data.dropna(inplace=True)

# Split data into features and target
X = data.drop('price_new', axis=1)
y = data['price_new']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=52)

# KNN Regressor
knn_regressor = KNeighborsRegressor(n_neighbors=5)
knn_regressor.fit(X_train, y_train)
knn_y_pred = knn_regressor.predict(X_test)

# Multiple Regression
multiple_regression = LinearRegression()
multiple_regression.fit(X_train, y_train)
multi_y_pred = multiple_regression.predict(X_test)

# Evaluation
def evaluate_model(test, pred):
    r2 = r2_score(test, pred)
    mse = mean_squared_error(test, pred)
    rmse = np.sqrt(mse)
    return r2, mse, rmse

knn_scores = evaluate_model(y_test, knn_y_pred)
multi_scores = evaluate_model(y_test, multi_y_pred)

print("KNN Regressor Scores:")
print("R2 Score:", knn_scores[0])
print("Mean Squared Error:", knn_scores[1])
print("Root Mean Squared Error:", knn_scores[2])

print("\nMultiple Regression Scores:")
print("R2 Score:", multi_scores[0])
print("Mean Squared Error:", multi_scores[1])
print("Root Mean Squared Error:", multi_scores[2])
 
```

```python
# Select the best model based on R2 score
best_model = "KNN Regressor" if knn_scores[0] > multi_scores[0] else "Multiple Regression"
print("Best Model:", best_model)
```
